# Streamlit Basics

Streamlit تسهل بناء تطبيقات ويب تفاعلية باستخدام بايثون فقط.

```python
import streamlit as st
```

```python
app_code = '''
import streamlit as st

st.set_page_config(page_title="Simple Demo", layout="centered")

st.title("Student Score Dashboard")
st.write("A minimal Streamlit example for learning the basics.")

student_name = st.text_input("Student name", "Ali")
score = st.slider("Score", 0, 100, 85)

if score >= 90:
    level = "Excellent"
elif score >= 75:
    level = "Very good"
else:
    level = "Needs improvement"

st.subheader("Result")
st.write(f"Student: {student_name}")
st.write(f"Score: {score}")
st.write(f"Level: {level}")
'''

print(app_code)
```

## طريقة التشغيل

ضع الكود في ملف باسم app.py ثم شغل الأمر التالي:
streamlit run app.py

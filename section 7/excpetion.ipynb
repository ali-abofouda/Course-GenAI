# Exception Handling

التعامل مع الاستثناءات يمنع توقف البرنامج بشكل مفاجئ ويجعل رسائل الخطأ أوضح للمستخدم والمطور.

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("You cannot divide by zero.")
```

```python
values = ["10", "20", "hello"]

for value in values:
    try:
        print(int(value))
    except ValueError:
        print(f"Cannot convert {value!r} to an integer.")
```

```python
try:
    number = int("42")
except ValueError:
    print("Conversion failed.")
else:
    print(f"Conversion succeeded: {number}")
finally:
    print("This block always runs.")
```

```python
def withdraw(balance, amount):
    if amount > balance:
        raise ValueError("Insufficient balance")
    return balance - amount

print(withdraw(1000, 250))
```

## ملخص

الأفضل التقاط نوع الخطأ المحدد بدل استخدام except عامة بدون داعٍ، حتى يبقى الكود مفهوماً ودقيقاً.

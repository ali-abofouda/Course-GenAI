# Loops in Python
This notebook introduces `for` and `while` loops with clean, incremental examples.

```python
# 1) Basic for loop
for i in range(10):
    print(i)
```

## Understanding range()
Use `range(stop)`, `range(start, stop)`, or `range(start, stop, step)`.

```python
# range(stop)
print(list(range(5)))
```

```python
# range(stop): starts from 0 to stop - 1
for i in range(5):
    print(i)

print("--------------------------------")
# range(start, stop): starts from 1 to stop - 1
for i in range(1, 5):
    print(i)

print("--------------------------------")
# range(start, stop, step): step by 2
for i in range(0, 10, 2):
    print(i)

print("--------------------------------")
# descending loop with negative step
for i in range(10, 0, -1):
    print(i)
```

## Looping over iterables
`for` works with strings, lists, tuples, and more.

```python
# string example
text = "Hello"
for ch in text:
    print(ch)
print("--------------------------------")
# list example
for item in [1, 2, 3, 4, 5]:
    print(item)
```

## While loops
`while` repeats as long as the condition is `True`.

```python
# while loop syntax
# while condition:
#     do_something()
#     update_condition()
```

```python
count = 0
while count < 5:
    print(f"count is {count}")
    count += 1
```

```python
# avoid infinite loops by updating the control variable
count = 0
while count % 2 == 0 and count < 10:
    print(f"count is {count}")
    count += 2
```

## Loop controls: break, continue, pass
Use these statements to change loop behavior.

```python
# break
for i in range(10):
    if i == 5:
        break
    print(i)
print("--------------------------------")
# continue
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# pass means: do nothing (placeholder)
for i in range(10):
    if i == 5:
        pass
    print(i)
```

## Nested loops
Useful for combinations, grids, and repeated structures.

```python
for i in range(3):
    for j in range(3):
        print(f"i: {i}, j: {j}")
```

## Practical example
Compute the sum from 1 to N using a loop.

```python
def calc_sum_n(n: int) -> int:
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

print(calc_sum_n(5))
```

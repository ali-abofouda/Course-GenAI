# Loops in Python

هذا الدفتر يشرح التكرار في بايثون باستخدام for و while بشكل مرتب وتدريجي.

## أهداف الدرس
- فهم فكرة التكرار ولماذا نستخدمه.
- التمييز بين for و while.
- كتابة أمثلة عملية قصيرة وواضحة.

## الفكرة الأساسية

نستخدم الحلقة عندما نريد تنفيذ نفس الفكرة أكثر من مرة.

- for مناسبة عندما نمر على عناصر موجودة أو عندما يكون عدد التكرارات واضحاً.
- while مناسبة عندما يعتمد التكرار على شرط يستمر ويتغير مع الوقت.

```python
# Basic for loop
for number in range(5):
    print(number)
```

```python
# Common range patterns
print(list(range(5)))
print(list(range(2, 8)))
print(list(range(0, 10, 2)))
```

```python
# Looping over strings and lists
name = "Ali"
scores = [90, 85, 98]

for letter in name:
    print(letter)

for score in scores:
    print(f"Score: {score}")
```

```python
# while loop example
count = 1
while count <= 5:
    print(f"Current count: {count}")
    count += 1
```

## ملخص

الحلقات تقلل تكرار الكود وتجعل الحلول أوضح. أهم نقطة هي اختيار النوع المناسب والانتباه جيداً لشرط الإيقاف في while.

```python
# tuples 
t = (1, 2, 3)
print(t[0])  # Output: 1
print(t[1])  # Output: 2
print(t[2])  # Output: 3
```

```python
# tuples are immutable
# t[0] = 10  # This will raise a TypeError: 'tuple
t[0] = 10  # This will raise a TypeError: 'tuple' object does not support item assignment
```

```python
mixed_tuple = (1, "hello", 3.14)
print(mixed_tuple)  # Output: (1, 'hello', 3.14
```

```python
# accessing tuple elements
print(mixed_tuple[0])  # Output: 1
print(mixed_tuple[1])  # Output: 'hello'
print(mixed_tuple[2])  # Output: 3.14
```

```python
# slicing tuples
print(mixed_tuple[0:2])  # Output: (1, 'hello')
print(mixed_tuple[1:])   # Output: ('hello', 3.14)
print(mixed_tuple[:2])   # Output: (1, 'hello')
```

```python
# tuples operations
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
# concatenation
tuple3 = tuple1 + tuple2
print(tuple3)  # Output: (1, 2, 3, 4, 5, 6)
# repetition
tuple4 = tuple1 * 2
print(tuple4)  # Output: (1, 2, 3, 1, 2, 3)
# length of a tuple
print(len(tuple1))  # Output: 3
```

```python
## immutable nature of tuples
# tuple1[0] = 10  # This will raise a TypeError: 'tuple' object does not support item assignment
# but list inside a tuple can be modified
tuple5 = (1, 2, [3, 4])
print(tuple5)  # Output: (1, 2, [3, 4])
tuple5[2][0] = 10
print(tuple5)  # Output: (1, 2, [10, 4
```

```python
tuple6 = (1, 2, 3 , [4 , 5])
print(tuple6)  # Output: (1, 2, 3, [4, 5])
tuple6[3][1] = 55
print(tuple6)  # Output: (1, 2, 3, [4, 55])
```

```python
## tuples methods
tuple7 = (1, 2, 3, 4, 5)
# count method means count the number of occurrences of a value in the tuple
print(tuple7.count(3))  # Output: 1
# index method means find the index of the first occurrence of a value in the tuple
print(tuple7.index(4))  # Output: 3
```

```python
# packing and unpacking tuples
# packing
packed_tuple = 1, 2, 3
print(packed_tuple)  # Output: (1, 2, 3)
# unpacking
a, b, c = packed_tuple
print(a)  # Output: 1
print(b)  # Output: 2
print(c)  # Output: 3
# unpacking with * operator
a, *b = packed_tuple
print(a)  # Output: 1
print(b)  # Output: [2, 3]
*a, b = packed_tuple
print(a)  # Output: [1, 2]
print(b)  # Output: 3
```

```python
# nested tuples
nested_tuple = (1, 2, (3, 4), [5, 6])
print(nested_tuple)  # Output: (1, 2, (3, 4), [5, 6])
print(nested_tuple[2])  # Output: (3, 4)
print(nested_tuple[3])  # Output: [5, 6]
print(type(nested_tuple[3]))  # Output: <class 'list'>

```

# Python Dictionaries

في هذا الدرس هنتعلم:
- يعني إيه `dict`
- إزاي نقرأ ونعدل البيانات
- أهم الدوال المستخدمة مع القواميس

```python
student = {
    "name": "Ali",
    "age": 22,
    "track": "AI"
}

print(student)
print(type(student))
```

## الوصول للبيانات والتعديل

- الوصول لقيمة: `student["name"]`
- الوصول الآمن: `student.get("key")`
- إضافة أو تعديل: `student["city"] = "Cairo"`

```python
print("Name:", student["name"])
print("Level:", student.get("level", "Not Found"))

student["city"] = "Cairo"
student["age"] = 23

print("Updated:", student)
print("Keys:", list(student.keys()))
print("Values:", list(student.values()))
print("Items:", list(student.items()))
```

## Loop على القاموس

نقدر نعمل loop على:
- المفاتيح
- القيم
- الزوج `(key, value)`

```python
for key in student:
    print("Key:", key)

for value in student.values():
    print("Value:", value)

for key, value in student.items():
    print(f"{key} -> {value}")
```

## دوال مفيدة

- `pop()` لحذف عنصر
- `update()` لتحديث بعدة عناصر
- `copy()` لنسخ القاموس

```python
profile = {"name": "Sara", "age": 21, "city": "Giza"}

removed_age = profile.pop("age")
print("Removed age:", removed_age)

profile.update({"track": "Data", "level": "Beginner"})
profile_copy = profile.copy()

print("Profile:", profile)
print("Copy:", profile_copy)
```

## تمرين سريع

اكتب قاموس اسمه `product` فيه:
- `name`
- `price`
- `stock`

ثم عدّل السعر واطبع كل العناصر باستخدام `items()`.

```python
# Solution
product = {
    "name": "Laptop",
    "price": 25000,
    "stock": 15
}

product["price"] = 23500

for key, value in product.items():
    print(f"{key}: {value}")
```

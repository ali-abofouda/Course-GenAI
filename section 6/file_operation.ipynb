```python
## file opeateions
with open("example.txt", "w") as file:  # This will open a file named "example.txt" in write mode. If the file does not exist, it will be created. If it already exists, it will be overwritten.
    file.write("Hello, World!")  # This will write the string "Hello, World!" to the file.
with open("example.txt", "r") as file:  # This will open the file named "example.txt" in read mode.
    content = file.read()  # This will read the entire contents of the file and store it in the variable content.
print(content)  # This will print the contents of the file, which should be "Hello, World!".
```

```python
with open("example.txt" , 'w') as file :
    file.write("This is a new line.")  # This will overwrite the existing content of the file with "This is a new line."
with open("example.txt", "r") as file:
    content = file.read()  # This will read the new content of the file and store it in the variable content.
print(content)  # This will print the new content of the file, which should be "This is a new line."
```

```python

## read line by line
with open("example.txt", "w") as file:
    file.write("Line 1\nLine 2\nLine 3")  # This will write three lines to the file, each separated by a newline character.     
with open("example.txt", "r") as file:
    for line in file:  # This will iterate through each line in the file.
        print(line.strip())  # This will print each line, removing any leading or trailing whitespace (including the newline character).
```

```python
## writeing multiple lines without overwriting
with open("example.txt", "a") as file:  # This will open the file in append mode, which allows you to add new content to the end of the file without overwriting the existing content.
    file.write("\nThis is an additional line.")  # This will add a new line to the end of the file. The "\n" at the beginning ensures that the new line starts on a new line in the file.  
with open("example.txt", "r") as file:
    content = file.read()  # This will read the entire contents of the file, including the newly added line, and store it in the variable content.
print(content)  # This will print the entire contents of the file, including the newly added line.
```

```python
## file .seek() and .tell()
with open("example.txt", "r") as file:
    print(file.tell())  # This will print the current position of the file pointer, which should be 0 at the beginning of the file.
    content = file.read(5)  # This will read the first 5 characters from the file and store it in the variable content.
    print(content)  # This will print the first 5 characters that were read from the file.
    print(file.tell())  # This will print the current position of the file pointer after reading, which should be 5.
    file.seek(0)  # This will move the file pointer back to the beginning of the file (position 0).
    print(file.tell())  # This will print the current position of the file pointer, which should be back to 0.
```

```python
## working with path and directories
import os
os.makedirs("new_directory/sub_directory", exist_ok=True)  # This will create a new directory named "new_directory" with a subdirectory named "sub_directory". The exist_ok=True parameter allows the function to ignore the error if the directories already exist.
with open("new_directory/sub_directory/example.txt", "w") as file:  # This will create a new file named "example.txt" inside the "sub_directory" of "new_directory" and open it in write mode.
    file.write("This file is located in a subdirectory.")  # This will write the string "This file is located in a subdirectory." to the newly created file.
with open("new_directory/sub_directory/example.txt", "r") as file:  # This  will open the file we just created in read mode.
    content = file.read()  # This will read the contents of the file and store it in the variable content.
print(content)  # This will print the contents of the file, which should be "This file is located in a subdirectory."
```

```python
items = os.listdir('.')
print(items)  # This will print a list of all files and directories in the current directory ('.').
```

```python
## joining paths
file_path = os.path.join("new_directory", "sub_directory", "example.txt")  # This will create a file path by joining the directory names and the file name. The resulting file_path will be "new_directory/sub_directory/example.txt" on Unix-like systems or "new_directory\sub_directory\example.txt" on Windows, depending on the operating system's path separator.
print(file_path)  # This will print the constructed file path, which should be "new_directory/sub_directory/example.txt" or "new_directory\sub_directory\example.txt" depending on the operating system.
```

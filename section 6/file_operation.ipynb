# File Operations in Python

هذا الدفتر يشرح القراءة والكتابة على الملفات باستخدام أسلوب مرتب وآمن.

```python
from pathlib import Path

file_path = Path("sample_notes.txt")
print(file_path)
```

```python
# Write to a file
with open(file_path, "w", encoding="utf-8") as file:
    file.write("Python makes file handling simple.
")
    file.write("This is the second line.
")
```

```python
# Read the file
with open(file_path, "r", encoding="utf-8") as file:
    content = file.read()

print(content)
```

```python
# Append and inspect line by line
with open(file_path, "a", encoding="utf-8") as file:
    file.write("This line was appended later.
")

with open(file_path, "r", encoding="utf-8") as file:
    for line_number, line in enumerate(file, start=1):
        print(f"{line_number}: {line.strip()}")
```

## لماذا نستخدم with؟

لأنها تضمن إغلاق الملف تلقائياً حتى لو حدث خطأ أثناء التنفيذ.

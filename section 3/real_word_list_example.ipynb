# Real World List Example

هذا المثال يربط فكرة القوائم بموقف عملي حتى تصبح فائدتها أوضح.

```python
tasks = [
    "Review Python basics",
    "Finish notebook organization",
    "Practice list comprehension",
    "Read about NLP"
]

print(tasks)
```

```python
# Add and remove tasks
tasks.append("Build a small project")
tasks.remove("Read about NLP")
print(tasks)
```

```python
# Number the tasks
for index, task in enumerate(tasks, start=1):
    print(f"{index}. {task}")
```

```python
# Filter tasks by keyword
python_tasks = [task for task in tasks if "Python" in task]
print(python_tasks)
```

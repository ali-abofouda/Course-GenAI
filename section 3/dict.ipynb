# Dictionaries in Python

القاموس dict يخزن البيانات على شكل مفتاح وقيمة، وهو مناسب جداً لتمثيل الأشياء الواقعية مثل المستخدم أو المنتج أو الإعدادات.

```python
student = {
    "name": "Ali",
    "age": 22,
    "track": "AI"
}

print(student["name"])
print(student["track"])
```

```python
# Safe access and update
print(student.get("city", "Not found"))
student["city"] = "Cairo"
student["age"] = 23
print(student)
```

```python
# Looping through a dictionary
for key, value in student.items():
    print(f"{key}: {value}")
```

```python
# Dictionary comprehension
numbers = [1, 2, 3, 4, 5]
squares = {number: number ** 2 for number in numbers}
print(squares)
```

## نقاط مهمة

- المفاتيح يجب أن تكون فريدة.
- get مفيدة عندما لا نضمن وجود المفتاح.
- dict اختيار ممتاز عندما يكون الوصول بالاسم أهم من الوصول بالموقع.

# Tuples in Python

ال tuple تشبه list لكنها غير قابلة للتعديل بعد الإنشاء، ولهذا فهي مناسبة للبيانات الثابتة.

```python
point = (10, 20)
colors = ("red", "green", "blue")

print(point)
print(colors)
print(point[0])
```

```python
# Tuple unpacking
name, age, track = ("Ali", 22, "AI")
print(name)
print(age)
print(track)
```

```python
# Returning multiple values from a function
def get_min_max(numbers):
    return min(numbers), max(numbers)

minimum, maximum = get_min_max([5, 10, 2, 18, 7])
print(minimum, maximum)
```

```python
# Single-item tuple
single_value = (42,)
print(type(single_value))
```

## ملخص

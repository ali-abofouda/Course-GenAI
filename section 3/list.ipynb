# Lists in Python

القائمة list هي أكثر بنية بيانات ستستخدمها غالباً في بايثون لأنها مرتبة وقابلة للتعديل بسهولة.

```python
fruits = ["apple", "banana", "orange", "mango"]

print(fruits)
print(fruits[0])
print(fruits[-1])
```

```python
# Slicing
print(fruits[1:3])
print(fruits[:2])
print(fruits[::2])
```

```python
# Update and loop
fruits.append("grape")
fruits.insert(1, "kiwi")
fruits.remove("orange")

for fruit in fruits:
    print(fruit)
```

```python
# List comprehension
numbers = [1, 2, 3, 4, 5]
squared_numbers = [number ** 2 for number in numbers]
even_numbers = [number for number in numbers if number % 2 == 0]

print(squared_numbers)
print(even_numbers)
```

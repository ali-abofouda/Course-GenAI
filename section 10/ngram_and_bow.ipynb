# N-Gram and Bag of Words

هذا الدفتر يشرح طريقتين أساسيتين لتحويل النصوص إلى تمثيل رقمي بسيط يمكن أن تتعامل معه خوارزميات التعلم الآلي.

```python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "I love learning natural language processing",
    "I love learning machine learning",
    "Natural language processing is fun"
]
```

```python
# Bag of Words
bow_vectorizer = CountVectorizer()
bow_matrix = bow_vectorizer.fit_transform(documents)

print(bow_vectorizer.get_feature_names_out())
print(bow_matrix.toarray())
```

```python
# Bigrams and mixed n-grams
bigram_vectorizer = CountVectorizer(ngram_range=(2, 2))
bigram_matrix = bigram_vectorizer.fit_transform(documents)
print(bigram_vectorizer.get_feature_names_out())
print(bigram_matrix.toarray())

mixed_vectorizer = CountVectorizer(ngram_range=(1, 2))
mixed_matrix = mixed_vectorizer.fit_transform(documents)
print(mixed_vectorizer.get_feature_names_out())
print(mixed_matrix.toarray())
```

## الفرق بسرعة

Bag of Words تهتم بعدد الكلمات المفردة، أما n-grams فتضيف معلومات عن تسلسل الكلمات معاً.

# TF-IDF

فكرة TF-IDF تعطي وزناً أكبر للكلمات المهمة داخل مستند وتقلل وزن الكلمات الشائعة جداً بين جميع المستندات.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "Python is great for data analysis",
    "Python is great for machine learning",
    "Data analysis uses pandas and numpy"
]
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
print(tfidf_matrix.toarray())
```

```python
feature_names = vectorizer.get_feature_names_out()
first_document_scores = tfidf_matrix.toarray()[0]

for word, score in zip(feature_names, first_document_scores):
    print(f"{word:<15} {score:.3f}")
```

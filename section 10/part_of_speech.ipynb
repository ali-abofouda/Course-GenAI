```python
# =========================
# Sample Text
# =========================
text = "Hello, my name is John. I am a student. I love programming and learning new things!"

# =========================
# Tokenization
# =========================
tokens = word_tokenize(text)
print("Tokens:", tokens)

sentences = sent_tokenize(text)
print("Sentences:", sentences)

# =========================
# Stemming (اختياري)
# =========================
ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in tokens]
print("Stemmed:", stemmed_words)

# =========================
# Lemmatization with POS
# =========================

lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(tag):
    """Convert POS tag to WordNet format"""
    if tag.startswith('J'):
        return 'a'  # adjective
    elif tag.startswith('V'):
        return 'v'  # verb
    elif tag.startswith('N'):
        return 'n'  # noun
    elif tag.startswith('R'):
        return 'r'  # adverb
    return 'n'

pos_tags = pos_tag(tokens)

lemmatized_words = [
    lemmatizer.lemmatize(word, get_wordnet_pos(tag))
    for word, tag in pos_tags
]

print("Lemmatized:", lemmatized_words)

# =========================
# Preprocessing Function (Production)
# =========================

def preprocess_text(text, lang='english'):
    """
    Full NLP preprocessing pipeline:
    - Lowercase
    - Remove punctuation
    - Tokenization
    - Stopwords removal
    - POS tagging
    - Lemmatization
    """

    # lowercase
    text = text.lower()

    # remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # tokenize
    tokens = word_tokenize(text)

    # remove stopwords
    stop_words = set(stopwords.words(lang))
    tokens = [word for word in tokens if word not in stop_words]

    # remove short words (optional)
    tokens = [word for word in tokens if len(word) > 2]

    # POS tagging
    pos_tags = pos_tag(tokens)

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    processed_tokens = [
        lemmatizer.lemmatize(word, get_wordnet_pos(tag))
        for word, tag in pos_tags
    ]

    return processed_tokens

# =========================
# Run Full Pipeline
# =========================

processed_text = preprocess_text(text)
print("Processed Text:", processed_text)

```

```python


# import necessary libraries
import nltk
import string

from sklearn.preprocessing import OneHotEncoder
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
from nltk import pos_tag


# download necessary resources
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
text = "Hello, my name is John. I am a student. I love programming."
# one-hot encoding
tokens = word_tokenize(text) # tokenize the text into words
print("Tokens:", tokens)
# one-hot encoding
encoder = OneHotEncoder() # create an instance of the OneHotEncoder
one_hot_encoded = encoder.fit_transform([[token] for token in tokens]) # fit and transform
print("One-Hot Encoded Shape:", one_hot_encoded.shape) # print the shape of the one-hot encoded matrix
print("One-Hot Encoded Matrix:\n", one_hot_encoded.toarray()) # print the one-hot encoded matrix as an array
```

```python
# advantage of one-hot encoding:
# - simplicity: easy to understand and implement
# - no assumptions: does not assume any relationship between categories
# - works well for small vocabularies

# disadvantage of one-hot encoding:
# - high dimensionality: can lead to sparse vectors and increased memory usage
# - does not capture semantic relationships: treats all words as independent and does not capture similarities between words
```

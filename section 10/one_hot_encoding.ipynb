# One-Hot Encoding for Words

التمثيل One-Hot يحول كل كلمة إلى متجه يحتوي على صفر في كل الخانات ما عدا خانة واحدة تمثل الكلمة نفسها.

```python
vocabulary = ["cat", "dog", "bird", "fish"]
word_to_index = {word: index for index, word in enumerate(vocabulary)}
print(word_to_index)
```

```python
def one_hot_encode(word, vocabulary):
    vector = [0] * len(vocabulary)
    if word in vocabulary:
        index = vocabulary.index(word)
        vector[index] = 1
    return vector

print(one_hot_encode("dog", vocabulary))
print(one_hot_encode("bird", vocabulary))
```

```python
sentence = ["cat", "fish", "dog"]
encoded_sentence = [one_hot_encode(word, vocabulary) for word in sentence]
print(encoded_sentence)
```

هذا التمثيل لا يعبر عن تشابه الكلمات ولا عن ترتيبها، كما أن حجمه يكبر بسرعة مع زيادة عدد الكلمات.

## 🔹 يعني إيه OHE؟

**One Hot Encoding (OHE)** هو أسلوب بنستخدمه علشان نحول البيانات الـ **Categorical** (نصوص أو فئات) لأرقام.

بيستخدم كتير في:

* Machine Learning
* Natural Language Processing

---

## 🔸 الفكرة الأساسية

بدل ما ندي كل فئة رقم (وده ممكن يسبب مشكلة)،
بنحول كل فئة لـ **Vector فيه 0 و 1 فقط** 👇

> كل فئة ليها مكان واحد = 1 والباقي = 0

---

## 🔸 مثال بسيط

عندك العمود ده:

```id="a1"
Color = [Red, Blue, Green]
```

### نعمل Encoding:

| Color | Red | Blue | Green |
| ----- | --- | ---- | ----- |
| Red   | 1   | 0    | 0     |
| Blue  | 0   | 1    | 0     |
| Green | 0   | 0    | 1     |

👉 كل قيمة بقت **Vector**

---

## 🔹 مثال في NLP

جملة:

```id="a2"
"I love AI"
```

### الكلمات:

```
[I, love, AI]
```

### OHE:

| Word | I | love | AI |
| ---- | - | ---- | -- |
| I    | 1 | 0    | 0  |
| love | 0 | 1    | 0  |
| AI   | 0 | 0    | 1  |

---

## 🔸 الفرق بين OHE و BoW 🔥

| Feature    | OHE         | BoW         |
| ---------- | ----------- | ----------- |
| بيعتمد على | وجود الكلمة | عدد التكرار |
| القيمة     | 0 أو 1      | 0,1,2,...   |
| الترتيب    | ❌ لا        | ❌ لا        |

👉 يعني:

* OHE = "هل الكلمة موجودة؟"
* BoW = "الكلمة ظهرت كام مرة؟"

---

## 🔸 مميزاته 👍

* بسيط جدًا
* مناسب للـ categorical data
* مفيش ترتيب وهمي بين القيم

---

## 🔸 عيوبه 👎

* ❌ بيكبر جدًا لو عدد الفئات كبير
* ❌ مبيفهمش معنى الكلمات
* ❌ Sparse (معظم القيم 0)

---

## 🔸 كود Python

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

data = np.array([["Red"], ["Blue"], ["Green"]])

encoder = OneHotEncoder()
encoded = encoder.fit_transform(data)

print(encoded.toarray())
```

## 🔥 الخلاصة

````python


# import necessary libraries
import nltk
import string

from sklearn.preprocessing import OneHotEncoder
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
from nltk import pos_tag


# download necessary resources
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger_eng')
````

````python
text = "Hello, my name is John. I am a student. I love programming."
# one-hot encoding
tokens = word_tokenize(text) # tokenize the text into words
print("Tokens:", tokens)
# one-hot encoding
encoder = OneHotEncoder() # create an instance of the OneHotEncoder
one_hot_encoded = encoder.fit_transform([[token] for token in tokens]) # fit and transform
print("One-Hot Encoded Shape:", one_hot_encoded.shape) # print the shape of the one-hot encoded matrix
print("One-Hot Encoded Matrix:\n", one_hot_encoded.toarray()) # print the one-hot encoded matrix as an array
````

````python
# advantage of one-hot encoding:
# - simplicity: easy to understand and implement
# - no assumptions: does not assume any relationship between categories
# - works well for small vocabularies

# disadvantage of one-hot encoding:
# - high dimensionality: can lead to sparse vectors and increased memory usage
# - does not capture semantic relationships: treats all words as independent and does not capture similarities between words
````

# Stemming and Lemmatization

كلتا الطريقتين تهدفان إلى تقليل الكلمات إلى صيغة أقرب إلى أصلها، لكن بينهما فرق مهم في الجودة.

```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Run these downloads once if needed.
# nltk.download("wordnet")
# nltk.download("omw-1.4")
```

```python
words = ["running", "studies", "better", "flying"]
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
print(lemmatized_words)

print(lemmatizer.lemmatize("running", pos="v"))
print(lemmatizer.lemmatize("better", pos="a"))
```

## الفرق السريع

Stemming أبسط وأسرع لكنه قد ينتج شكلاً غير صحيح لغوياً، بينما Lemmatization أدق غالباً عندما يهمنا الشكل الحقيقي للكلمة.

# Stemming و Lemmatization في NLP

## الفرق بين Stemming و Lemmatization

الفرق بين Stemming و Lemmatization في الـ NLP بسيط لكن مهم 👇

### ✂️ Stemming
- بيقص الكلمة بشكل عشوائي للوصول لجذرها
- مش بيعتمد على معنى الكلمة
- سريع جدًا لكن أحيانًا بيدي كلمات غلط
- مثال:
  - playing → play  
  - studies → studi ❌

📌 زي كأنك بتقص الكلمة بأي شكل عشان تختصرها

### 🧠 Lemmatization
- بيرجع الكلمة لأصلها الصحيح (Lemma)
- بيعتمد على القواعس اللغوية والمعنى
- أبطأ شوية لكنه أدق
- مثال:
  - playing → play  
  - studies → study ✅  
  - better → good ✅

📌 هنا بيفهم اللغة مش بس بيقص

### ⚖️ الفرق باختصار

| Feature | Stemming ✂️ | Lemmatization 🧠 |
|---------|-----------|-----------------|
| الدقة | أقل | أعلى |
| السرعة | أسرع | أبطأ |
| بيعتمد على معنى | ❌ لا | ✅ نعم |
| الناتج | ممكن يكون غلط | صحيح لغويًا |

### 🎯 تستخدم إمتى؟
- لو عايز سرعة ومش فارق معاك الدقة → **Stemming**
- لو عايز نتيجة مظبوطة (زي search أو AI models) → **Lemmatization**

## ملخص الاختلافات الرئيسية

**Stemming:** طريقة بسيطة وسريعة تقطع الكلمات بشكل عشوائي، قد تسبب مشاكل لكن تعطي نتائج أسرع.

**Lemmatization:** طريقة متطورة تستخدم قواعس اللغة والقاموس، أبطأ لكن أدق في النتائج وتعتبر الخيار الأفضل للتطبيقات التي تحتاج دقة عالية.

```python
# ============================================================
# القسم 1: استيراد المكتبات والموارد اللازمة
# ============================================================

# تحميل الموارد اللازمة (يجب تنفيذها مرة واحدة فقط)
import nltk
nltk.download('wordnet')

from nltk.stem import PorterStemmer, SnowballStemmer, WordNetLemmatizer
import re

print("✅ تم تحميل المكتبات بنجاح")
```

```python
# ============================================================
# 1️⃣ Porter Stemmer - استخدام ألغوريتمية بسيطة للقطع
# ============================================================

stemmer = PorterStemmer()
words = ["running", "ran", "runs", "easily", "fairly"]

# تطبيق Porter Stemming
porter_stemmed = [stemmer.stem(word) for word in words]

print("=" * 50)
print("1️⃣ PORTER STEMMER")
print("=" * 50)
print(f"الكلمات الأصلية:    {words}")
print(f"بعد Porter Stemming: {porter_stemmed}")
print("\n✅ المميزات:")
print("   - سريع جدًا")
print("   - بسيط وسهل الاستخدام")
print("\n❌ العيوب:")
print("   - قد تعطي نتائج غير منطقية (مثل 'studi' بدل 'study')")
print("   - لا تعتمد على المعنى")
print("=" * 50)
```

```python
# ============================================================
# 2️⃣ Regex Stemmer - استخدام التعبيرات النمطية
# ============================================================

def regex_stemmer(word):
    """
    طريقة بسيطة جدًا - تزيل الصيغ الشائعة من نهاية الكلمة
    """
    word = re.sub(r'(ing|ed|s)$', '', word)
    return word

# تطبيق Regex Stemming
regex_stemmed = [regex_stemmer(word) for word in words]

print("=" * 50)
print("2️⃣ REGEX STEMMER")
print("=" * 50)
print(f"الكلمات الأصلية:    {words}")
print(f"بعد Regex Stemming: {regex_stemmed}")
print("\n✅ المميزات:")
print("   - سريع جدًا")
print("   - يمكن تخصيصه بسهولة")
print("\n❌ العيوب:")
print("   - بسيط وقد يعطي نتائج خاطئة")
print("   - لا يفهم السياق على الإطلاق")
print("   - مثال: 'running' تصبح 'runn' بدل 'run'")
print("=" * 50)
```

```python
# ============================================================
# 3️⃣ Snowball Stemmer - تطور من Porter مع دعم لغات متعددة
# ============================================================

snowball_stemmer = SnowballStemmer("english")

# تطبيق Snowball Stemming
snowball_stemmed = [snowball_stemmer.stem(word) for word in words]

print("=" * 50)
print("3️⃣ SNOWBALL STEMMER")
print("=" * 50)
print(f"الكلمات الأصلية:       {words}")
print(f"بعد Snowball Stemming: {snowball_stemmed}")
print("\n✅ المميزات:")
print("   - أفضل من Porter Stemmer")
print("   - يدعم لغات متعددة")
print("   - أكثر قوة في التقطيع")
print("\n❌ العيوب:")
print("   - أبطأ قليلاً من Porter")
print("   - قد تزال أجزاء مهمة من الكلمة")
print("=" * 50)
```

```python
# ============================================================
# 4️⃣ WordNet Lemmatizer - استخدام القواموس والقواعد اللغوية
# ============================================================

lemmatizer = WordNetLemmatizer()

# تطبيق Lemmatization
lemmatized = [lemmatizer.lemmatize(word, pos='v') for word in words]

print("=" * 50)
print("4️⃣ WORDNET LEMMATIZER")
print("=" * 50)
print(f"الكلمات الأصلية:     {words}")
print(f"بعد Lemmatization:  {lemmatized}")
print("\n✅ المميزات:")
print("   - تعطي نتائج صحيحة لغويًا")
print("   - تفهم معنى الكلمة من السياق")
print("   - النتائج منطقية دائمًا")
print("\n❌ العيوب:")
print("   - أبطأ من الطرق الأخرى")
print("   - تحتاج موارد أكثر")
print("   - قد لا تعرف كلمات جديدة")
print("=" * 50)
```

## الملخص والتوصيات

### 📌 الخلاصة:
1. **Stemming** أسرع لكن أقل دقة - استخدمها عندما السرعة أهم من الدقة
2. **Lemmatization** أبطأ لكن أدق - استخدمها في التطبيقات التي تحتاج دقة عالية
3. **Regex Stemming** بسيط جدًا - لا توصى به للتطبيقات الاحترافية
4. **Snowball Stemmer** أفضل من Porter - خيار جيد للـ stemming

### 🎯 التطبيقات الفعلية:
- **محركات البحث**: lemmatization
- **تحليل المشاعر**: lemmatization  
- **تصنيف النصوص**: قد يكفي stemming
- **معالجة اللغة الطبيعية (NLP)**: lemmatization

```python
# ============================================================
# القسم الإضافي: تمرين عملي - معالجة جملة كاملة
# ============================================================

# نص عينة
sample_text = "The students were running and playing, they enjoyed studying linguistics easily"

# تقسيم النص إلى كلمات
words_from_text = sample_text.lower().split()

print("=" * 70)
print("📝 معالجة جملة كاملة باستخدام جميع الطرق")
print("=" * 70)
print(f"النص الأصلي: {sample_text}\n")

# Regex Stemming
regex_results = [regex_stemmer(word.replace('.', '')) for word in words_from_text]

# Porter Stemmer
porter_results = [stemmer.stem(word.replace('.', '')) for word in words_from_text]

# Snowball Stemmer
snowball_results = [snowball_stemmer.stem(word.replace('.', '')) for word in words_from_text]

# WordNet Lemmatizer
lemma_results = [lemmatizer.lemmatize(word.replace('.', ''), pos='v') for word in words_from_text]

# جدول المقارنة
print(f"{'الكلمة':<12} | {'Regex':<12} | {'Porter':<12} | {'Snowball':<12} | {'Lemma':<12}")
print("-" * 70)
for i, word in enumerate(words_from_text):
    clean_word = word.replace('.', '')
    print(f"{clean_word:<12} | {regex_results[i]:<12} | {porter_results[i]:<12} | {snowball_results[i]:<12} | {lemma_results[i]:<12}")
print("=" * 70)
```

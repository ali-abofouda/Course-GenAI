# Text Preprocessing and Stop Words

تنظيف النصوص خطوة أساسية قبل بناء أي نموذج NLP، لأن النص الخام غالباً يحتوي على ضوضاء لا نحتاجها مباشرة.

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Run these downloads once if needed.
# nltk.download("punkt")
# nltk.download("punkt_tab")
# nltk.download("stopwords")
```

```python
text = "This is a simple example, showing how text preprocessing works in NLP!"
clean_text = re.sub(r"[^a-zA-Z\s]", "", text).lower()
print(clean_text)
```

```python
tokens = word_tokenize(clean_text)
print(tokens)

stop_words = set(stopwords.words("english"))
filtered_tokens = [token for token in tokens if token not in stop_words]
print(filtered_tokens)
```

```python
# text preprocessing remove stop words, punctuation
import re
import string
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def remove_stop_words(text , language="english"):
    """this function removes stop words and punctuation from the input text"""
    # Remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))
    
    # Remove stop words
    words = text.split()
    stop_words = stopwords.words(language)
    cleaned_words = [word for word in words if word.lower() not in stop_words]
    cleaned_text = " ".join(cleaned_words)
    return cleaned_text

# Example usage
english_text = "This is an example sentence, with some stop words and punctuation!"
arabic_text = "هذه جملة مثال، تحتوي على بعض الكلمات الوقفية وعلامات الترقيم!"


cleaned_arabic_text = remove_stop_words(arabic_text, language="arabic")
print("Original Arabic Text:", arabic_text)
print("Cleaned Arabic Text:", cleaned_arabic_text)

print("*"*20)

cleaned_english_text = remove_stop_words(english_text, language="english")
print("original English Text:", english_text)
print("Cleaned English Text:", cleaned_english_text)
```

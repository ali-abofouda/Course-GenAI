**NLTK** اختصار لـ **Natural Language Toolkit**
وهي مكتبة في بايثون بتستخدم في **معالجة اللغة الطبيعية (NLP)**.

## 💡 ببساطة بتعمل إيه؟

* تقسيم النص لكلمات أو جمل (Tokenization)
* تنظيف النص (حذف stopwords)
* تحويل الكلمات لأصلها (Stemming / Lemmatization)
* تحليل المشاعر (Sentiment Analysis)
* التعامل مع النصوص وتحليلها بشكل عام

## ⚙️ مثال سريع:

```python
import nltk
from nltk.tokenize import word_tokenize

text = "I love AI and NLP"
tokens = word_tokenize(text)

print(tokens)
```

📌 الناتج:

```
['I', 'love', 'AI', 'and', 'NLP']
```

````python
# nltk is a popular library for natural language processing in Python.
# It provides tools for tokenization, stemming, tagging, parsing, and more.
# You can use it to analyze and manipulate text data in various ways.

import nltk
from nltk.tokenize import word_tokenize
# Download the necessary resources (only need to do this once)

nltk.download('punkt')
nltk.download('punkt_tab')

````

````python
text = "Hello my name is ali and i will be reaching in 2030"
corpus = """Hello my name is ali and i will be reaching in 2030
I am a software engineer and i will be working on AI projects in the future
I am excited to learn more about natural language processing and machine learning
I hope to contribute to the field of AI and make a positive impact on society
I am looking forward to the advancements in AI and how it will shape our world
I am also interested in the ethical implications of AI and how we can ensure it is used responsibly"""

corpus= corpus.split("\n")

print(corpus)
````

````python
# Tokenize the text into sentences  
from nltk.tokenize import sent_tokenize
corpus_tokenized = [sent_tokenize(sentence) for sentence in corpus]


print(corpus_tokenized)
````

````python
# word tokenize the text into words
from nltk.tokenize import word_tokenize
corpus_tokenized = [word_tokenize(sentence) for sentence in corpus]
print(corpus_tokenized)
````

# Polymorphism

تعدد الأشكال يعني أن نفس الواجهة أو نفس اسم الدالة يمكن أن يعمل بطرق مختلفة حسب نوع الكائن.

```python
class Bird:
    def move(self):
        print("Bird is flying")

class Fish:
    def move(self):
        print("Fish is swimming")

class Horse:
    def move(self):
        print("Horse is running")

animals = [Bird(), Fish(), Horse()]
for animal in animals:
    animal.move()
```

```python
def describe_payment(payment_method, amount):
    payment_method.pay(amount)

class CashPayment:
    def pay(self, amount):
        print(f"Paid {amount} in cash")

class WalletPayment:
    def pay(self, amount):
        print(f"Paid {amount} using a wallet app")

describe_payment(CashPayment(), 100)
describe_payment(WalletPayment(), 150)
```

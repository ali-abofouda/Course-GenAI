```python
## polymorphism
class Car:
    def __init__(self , name , color):
        self.name = name 
        self.color = color
    def info(self):
        print(f"the type of car is => {self.name} and the color is => {self.color}")
class Bike:
    def __init__(self , name , color):
        self.name = name 
        self.color = color
    def info(self):
        print(f"the type of bike is => {self.name} and the color is => {self.color}")
car1 = Car('bmw' , 'red')
bike1 = Bike('honda' , 'blue')
car1.info() # the type of car is => bmw and the color is => red
bike1.info() # the type of bike is => honda and the color is => blue
## so both classes have the same method name but they do different things this is called polymorphism 
```

```python
for vehicle in (car1 , bike1):
    vehicle.info()
```

```python
class Shape:
    def area(self):
        pass
class Square(Shape):
    def __init__(self , side):
        self.side = side
    def area(self):
        return self.side * self.side
class Circle(Shape):
    def __init__(self , radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius * self.radius
square = Square(4)
circle = Circle(3)
s = Shape() # this is an abstract class we cannot create an instance of it because it has a method that is not implemented
print(square.area()) # 16
print(circle.area()) # 28.26
```

```python
# polymorphism with abstract classes
from abc import ABC , abstractmethod # abstract base class
## abstract class is a class that cannot be instantiated and it is used as a base class for other classes
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Square(Shape):
    def __init__(self , side):
        self.side = side
    def area(self):
        return self.side * self.side
class Circle(Shape):
    def __init__(self , radius):
        self.radius = radius
    def area(self):
        return 3.14 * self.radius * self.radius
square = Square(4)
circle = Circle(3)
# s = Shape() # this will raise an error because we cannot instantiate an abstract class
print(square.area()) # 16
print(circle.area()) # 28.26
```

```python
## classes and objects in python
### a class is a blueprint for creating objects. An object is an instance of a class. A class can have attributes (data) and methods (functions) that operate on the data.
### to define a class in python, we use the class keyword followed by the class name and a colon. The class name should be in CamelCase.
### to create an object of a class, we call the class as if it were a function, passing any arguments that the __init__ method requires. The __init__ method is a special method that is called when an object is created. It is used to initialize the attributes of the object.
### here is an example of a simple class in python:

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")
# create an object of the Person class
person1 = Person("Alice", 30)
# call the greet method
person1.greet()
### in this example, we defined a class called Person with an __init__ method that initializes the name and age attributes. We also defined a greet method that prints a greeting message. We then created an object of the Person class called person1 and called the greet method on that object.
```

```python
class Car:
    def __init__(self , name , color):
        self.name = name 
        self.color = color
    def info(self):
        print(f"the type of car is => {self.name} and the color is => {self.color}")

car1 = Car('bmw' , 'red')

car1.info()
```

```python
## modling a bank account using classes and objects
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New balance is {self.balance}.")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds.")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance is {self.balance}.")
# create an object of the BankAccount class
account1 = BankAccount("123456789", 1000)
# deposit money
account1.deposit(500)
# withdraw money
account1.withdraw(200)
# withdraw more than the balance
account1.withdraw(1500)

```

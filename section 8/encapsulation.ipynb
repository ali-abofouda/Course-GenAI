```python
## encapsulation is the process of hiding the internal details of an object and only exposing a public interface to interact with it. This is achieved by using access modifiers such as private, protected, and public.
## In Python, we can use the following conventions to indicate the level of access for class members:
## 1. Public members: These are accessible from anywhere and are defined without any special prefix. For example:
class MyClass_Public:
    def __init__(self, value):
        self.value = value  # public member

    def public_method(self):
        return f"The value is {self.value}"
## 2. Protected members: These are intended to be accessed only within the class and its subclasses. They are defined with a single underscore prefix. For example:
class MyClass_Protected:
    def __init__(self, value):
        self._value = value  # protected member

    def _protected_method(self):
        return f"The protected value is {self._value}"
## 3. Private members: These are intended to be accessed only within the class and are defined with a double underscore prefix. For example:
class MyClass_Private:
    def __init__(self, value):
        self.__value = value  # private member

    def __private_method(self):
        return f"The private value is {self.__value}"
    
obj1 = MyClass_Public(10)
print(obj1.public_method())  # Output: The value is 10

obj2 = MyClass_Protected(20)
print(obj2._protected_method())  # Output: The protected value is 20

obj3 = MyClass_Private(30)
print(obj3.__private_method())  # This will raise an AttributeError because __private_method is not accessible outside the class
```

```python
class Test:
    def __init__(self):
        self.public = 1
        self._protected = 2
        self.__private = 3

obj = Test()

print(obj.public)        # ✅
print(obj._protected)    # ⚠️ ينفع بس مش صح
# print(obj.__private)   # ❌ error
```

```python
## Getters and setters
## Getters and setters are methods that allow you to access and modify the private members of a class. They provide a way to control how the private members are accessed and modified, ensuring that the internal state of the object is maintained correctly.
class Test:
    def __init__(self):
        self.__private = 3

    def get_private(self):
        return self.__private

    def set_private(self, value):
        self.__private = value

obj = Test()
print(obj.get_private())  # Output: 3
obj.set_private(10)
print(obj.get_private())  # Output: 10
```

```python
## Inheritance from a class with protected members 
class Parent:
    def __init__(self):
        self._protected = 2
class Child(Parent):
    def access_protected(self):
        return self._protected
child = Child()
print(child.access_protected())  # Output: 2

## anoterh example of inheritance from a class with protected members
class Person:
    def __init__(self, name):
        self._name = name  # protected member
class Employee(Person):
    def get_name(self):
        return self._name
employee = Employee("Alice")
print(employee.get_name())  # Output: Alice
```

```python
class BankAccount:
    def __init__(self , serial_number , balance):
        self.__serial_number = serial_number
        self.__balance = balance
    
    def get_balance(self):
        return self.__balance
    
    def set_balance(self , new_balance):
        self.__balance = new_balance



BA1 = BankAccount(12345 , 5000)
print(BA1.get_balance()) # 5000
BA1.set_balance(300) 
print(BA1.get_balance())



    
```

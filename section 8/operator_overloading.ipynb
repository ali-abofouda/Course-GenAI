# Operator Overloading

تحميل المعاملات يعني تعريف سلوك خاص للعمليات مثل الجمع والمقارنة عند استخدامها مع كائناتنا.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __repr__(self):
        return f"Vector(x={self.x}, y={self.y})"

first = Vector(3, 4)
second = Vector(1, 2)

print(first + second)
print(first - second)
```

```python
class Money:
    def __init__(self, amount):
        self.amount = amount

    def __eq__(self, other):
        return self.amount == other.amount

wallet_1 = Money(100)
wallet_2 = Money(100)
print(wallet_1 == wallet_2)
```

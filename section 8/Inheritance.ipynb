```python
## inheritance in python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass
class Dog(Animal):
    def speak(self):
        return "Woof!"
class Cat(Animal):
    def speak(self):
        return "Meow!"
dog = Dog("Buddy")
cat = Cat("Whiskers")
print(f"{dog.name} says {dog.speak()}")
print(f"{cat.name} says {cat.speak()}")
```

```python
class Person:
    def __init__(self , name):
        self.name = name


class Employee(Person):
    def __init__(self , name , salary):
        super().__init__(name) # call the parent class constructor
        self.salary = salary
employee1 = Employee("Alice" , 50000)
print(f"Employee Name: {employee1.name}, Salary: {employee1.salary}")
```

```python
class Vehicle:
    def __init__(self , make , model):
        self.make = make
        self.model = model
class Car(Vehicle):
    def __init__(self , make , model , color):
        super().__init__(make , model)
        self.color = color
    def info(self):
        print(f"the type of car is => {self.make} {self.model} and the color is => {self.color}")
car1 = Car('bmw' , 'x5' , 'red')
car1.info()
```

```python
## multiple inheritance
class A:
    def method_a(self):
        return "Method A from class A"
class B:
    def method_b(self):
        return "Method B from class B"
class C(A , B):
    def method_c(self):
        return "Method C from class C"
c_instance = C()
print(c_instance.method_a())  # Output: Method A from class A
print(c_instance.method_b())  # Output: Method B from class B
print(c_instance.method_c())  # Output: Method C from class C
```

```python
class Person:
    def __init__(self, name, **kwargs):
        super().__init__(**kwargs)
        self.name = name


class Employee(Person):
    def __init__(self, salary, **kwargs):
        super().__init__(**kwargs)
        self.salary = salary


class Student(Person):
    def __init__(self, grade, **kwargs):
        super().__init__(**kwargs)
        self.grade = grade


class TeachingAssistant(Employee, Student):
    def __init__(self, name, salary, grade):
        super().__init__(name=name, salary=salary, grade=grade)
ta = TeachingAssistant(name="Alice", salary=50000, grade="A")
print(f"Name: {ta.name}, Salary: {ta.salary}, Grade: {ta.grade}")
```

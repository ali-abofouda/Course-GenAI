# Inheritance

الوراثة تسمح ببناء فئة جديدة اعتماداً على فئة موجودة، وهذا يساعدنا على إعادة استخدام السلوك المشترك.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("Animal sound")

class Dog(Animal):
    def speak(self):
        print(f"{self.name} says woof")

class Cat(Animal):
    def speak(self):
        print(f"{self.name} says meow")

dog = Dog("Max")
cat = Cat("Luna")

dog.speak()
cat.speak()
```

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def move(self):
        print("Vehicle is moving")

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

    def details(self):
        print(f"{self.brand} {self.model}")

car = Car("Toyota", "Corolla")
car.move()
car.details()
```

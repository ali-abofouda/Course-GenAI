# Magic Methods

ال magic methods هي دوال خاصة تسمح للكائنات بالتعامل مع بايثون بشكل طبيعي في الطباعة والمقارنة والطول وغيرها.

```python
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __str__(self):
        return f"Book(title={self.title}, pages={self.pages})"

    def __len__(self):
        return self.pages

book = Book("Python Basics", 180)
print(book)
print(len(book))
```

```python
class Counter:
    def __init__(self, start):
        self.start = start

    def __repr__(self):
        return f"Counter(start={self.start})"

counter = Counter(10)
print(counter)
```

## ملخص

هذه الدوال تجعل كائناتنا أقرب في الاستخدام إلى الأنواع المدمجة داخل بايثون.

# Import in Python

الاستيراد يسمح لنا باستخدام أكواد جاهزة من مكتبات بايثون القياسية أو من مكتبات خارجية أو من ملفات داخل المشروع.

```python
import math

print(math.sqrt(25))
print(math.pi)
```

```python
from random import randint
print(randint(1, 10))
```

```python
import statistics as stats
scores = [85, 90, 88, 92]
print(stats.mean(scores))
```

```python
import pandas as pd

data = pd.DataFrame(
    {
        "name": ["Ali", "Mona"],
        "score": [95, 98]
    }
)

print(data)
```

## ملخص

- import module يجعل المصدر واضحاً.
- from module import name يقلل الكتابة.
- alias مثل pd و np شائع عندما يكون الاسم معروفاً بين المطورين.

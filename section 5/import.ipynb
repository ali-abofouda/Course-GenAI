```python
# import module means to bring in a module (a file containing Python code) into the current namespace so that you can use its functions, classes, and variables. The syntax for importing a module is as follows:
# import module_name
import math  # This will import the math module, which provides mathematical functions and constants.
print(math.sqrt(16))  # This will use the sqrt function from the math module to calculate the square root of 16. The output will be 4.0.
```

```python
import numpy as np  # This will import the numpy module and give it the alias np, which is commonly used in the Python community for numerical operations.
print(np.array([1, 2, 3]))  # This will create a numpy array from the list [1, 2, 3] and print it. The output will be [1 2 3].
```

```python
from math import * # This will import all functions and variables from the math module directly into the current namespace, allowing you to use them without the math. prefix.
print(sin(pi / 2))  # This will use the sin function and pi constant from the math module to calculate the sine of pi/2. The output will be 1.0.
```

```python
## standard library 
import os  # This will import the os module, which provides functions for interacting with the operating system.
print(os.getcwd())  # This will print the current working directory.
os.makedirs("new_folder", exist_ok=True)  # This will create a new folder named "new_folder" in the current working directory. If the folder already exists, it will not raise an error due to exist_ok=True.
import sys  # This will import the sys module, which provides access to some variables used or maintained by the Python interpreter and to functions that interact strongly with the interpreter.
print(sys.version)  # This will print the Python version.
  
import random  # This will import the random module, which provides functions for generating random numbers.
print(random.randint(1, 10))  # This will print a random integer between 1 and 10 (inclusive).
print(random.choice(["Real Madrid", "Barcelona", "Bayern Munich"]))  # This will print a random element from the list.
```

```python
## high level operations on files and directories
import shutil  # This will import the shutil module, which provides high-level operations on files and directories.
shutil.copy("source.txt", "destination.txt")  # This will copy the file named "source.txt" to a new file named "destination.txt". If "destination.txt" already exists, it will be overwritten.
shutil.move("old_folder", "new_folder")  # This will move the folder named "old_folder" to a new location named "new_folder". If "new_folder" already exists, it will be overwritten.
shutil.rmtree("folder_to_delete")  # This will delete the folder named "folder_to_delete" and all of its contents. Use with caution, as this operation is irreversible. 
```

```python
## data serialization
import json  # This will import the json module, which provides functions for working with JSON data
data = {"name": "Alice", "age": 30, "city": "New York"}  # This is a Python dictionary that we want to serialize to JSON.
json_string = json.dumps(data)  # This will convert the Python dictionary into a JSON-form
```

```python
type(json_string) # This will print the type of the variable json_string, which is <class 'str'>.
json_string
```

```python
import csv 
with open("example.csv" , 'w' , newline='') as file :
    writer = csv.writer(file)
    writer.writerow(['name' , 'age'])
    writer.writerow(["ali" ,23])

with open("example.csv" , 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
import datetime  # This will import the datetime module, which provides classes for manipulating dates and times.
print(datetime.datetime.now())  # This will print the current date and time.  
print(datetime.date.today())  # This will print the current date.
print(datetime.timedelta(days=7))  # This will print a timedelta object representing a duration of 7 days. You can use this to perform date arithmetic, such as adding it to a date to get a new date.
```

```python
## regular expressions
import re  # This will import the re module, which provides functions for working with regular expressions
pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'  # This is a regular expression pattern for matching email addresses.
text = "Please contact us at john@example.com or jane@example.com"
emails = re.findall(pattern, text)  # This will find all occurrences of the email pattern in the text and return them as a list.
print(emails)  # This will print the list of email addresses found in the text.
```

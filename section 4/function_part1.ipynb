## أول مثال: دالة بدون Parameters

الدالة بتجمع كود متكرر في مكان واحد.

```python
def say_hello():
    print("Hello from function")

say_hello()
say_hello()
```

## دالة بـ Parameters

بنمرر قيم للدالة وقت الاستدعاء.

```python
def greet(name):
    print(f"Hello, {name}")

greet("Ali")
greet("Mona")
```

## Return Value

الدالة ممكن ترجع قيمة نستخدمها بعد كده.

```python
def add_numbers(a, b):
    return a + b

result = add_numbers(10, 5)
print("Result:", result)
```

## Default Parameter

نقدر نحط قيمة افتراضية للـ parameter.

```python
def power(base, exp=2):
    return base ** exp

print(power(5))
print(power(2, 3))
```

## تمرين

اكتب دالة اسمها `is_even(num)` ترجع:
- `True` لو الرقم زوجي
- `False` لو الرقم فردي

```python
# Solution
def is_even(num):
    return num % 2 == 0

print(is_even(4))
print(is_even(7))
```

الدوال تساعدنا على تقسيم البرنامج إلى أجزاء صغيرة قابلة لإعادة الاستخدام وأسهل في الفهم والاختبار.

```python
def greet():
    print("Hello from Python")

greet()
```

```python
def greet_user(name):
    print(f"Hello, {name}")

greet_user("Ali")
greet_user("Sara")
```

```python
def add_numbers(first, second):
    return first + second

result = add_numbers(10, 15)
print(result)
```

```python
def calculate_average(numbers):
    if not numbers:
        return 0
    return sum(numbers) / len(numbers)

print(calculate_average([90, 85, 100]))
```
